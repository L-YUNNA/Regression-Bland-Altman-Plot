```python
def get_ci(y_true, y_pred, metric, num_iter):
    rng = np.random.RandomState(seed=1234)
    idx = np.arange(y_true.shape[0])
    
    values = []
    for i in range(num_iter):
        resampled_idx = rng.choice(idx, size=idx.shape[0], replace=True)
        resampled_y_true = y_true(resampled_idx)
        resampled_y_pred = y_pred(resampled_idx)
        
        if metric=='MAE':
            value = mean_absolute_error(resampled_y_true, resampled_y_pred)
        elif metric=='RMSE':
            value = mean_squared_error(resampled_y_true, resampled_y_pred, squared=False)
        values.append(value)
    return np.percentile(values, 2.5), np.percentile(values, 97.5)
```

### Bland-atlman Plot

#### - Bland-altman with linear regression line

```python
means = []
diffs = []

for i in range(len(y_true)):
    avg = (y_true[i]+y_pred[i])/2
    means.append(avg)
    
    diff = y_true[i]-y_pred[i]
    diffs.append(diff)
```
